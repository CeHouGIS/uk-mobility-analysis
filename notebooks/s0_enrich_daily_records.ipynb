```python
import pyreadr

result = pyreadr.read_r('/nas/houce/UK_mobility/device_aid_keep_England.Rdata')
# result is a dictionary where keys are the names of objects loaded from the RData file
# For example, to get the first object:
df = next(iter(result.values()))
df = df.rename(columns={'device_aid_keep': 'device_aid'})
```

```python
import pyreadr

result0 = pyreadr.read_r("/nas/houce/UK_mobility/2022 March/UK_2022_0301.RData")
df0 = next(iter(result0.values()))
df0 = df0.reset_index(drop=True)
df_merge_ini = df.merge(df0, on='device_aid', how='inner')
```

```python
file_name = "/nas/houce/UK_mobility/2020 Feb/UK_2020_0201.RData"
filename_day_info = file_name.split('/')[-1].split('.')[0].split('_')
day_info = filename_day_info[1] + filename_day_info[2]
```

```python
df_merge = df0.copy()
df_merge['day'] = day_info
df_merge['location'] = df_merge.apply(lambda row: f"{row['latitude']}_{row['longitude']}", axis=1)
df_merge['hour_tem'] = df_merge['hour'].apply(lambda x: int(x.split(' ')[-1]))
df_merge.drop(columns=['hour'], inplace=True)
df_merge = df_merge.rename(columns={'hour_tem': 'showed_exact_hours'})
```

```python
df_merge['hour_count'] = df_merge.groupby(['device_aid', 'showed_exact_hours'])['showed_exact_hours'].transform('count')
df_merge['place_count'] = df_merge.groupby(['device_aid', 'location', 'showed_exact_hours'])['showed_exact_hours'].transform('count')
df_merge['single_location_hour'] = df_merge['place_count'] / df_merge['hour_count']
df_merge['showed_hours_today'] = df_merge.groupby(['device_aid', 'day'])['showed_exact_hours'].transform('nunique')
df_merge['duration_location_hour'] = df_merge.groupby(['device_aid', 'location'])['single_location_hour'].transform('sum')
df_merge_tem = df_merge.drop(columns = ['showed_exact_hours', 'place_count', 'hour_count', 'single_location_hour']).drop_duplicates().reset_index(drop=True)
```

```python
df_merge_tem.to_csv('/nas/houce/UK_mobility/processed_data/UK_2020_0201_processed.csv', index=False)
```
