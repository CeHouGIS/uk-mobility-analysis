```python
# import plotly.express as px
import pandas as pd

data = pd.read_csv("/nas/houce/UK_mobility/fig_output/final_analysis_data.csv")
# data = pd.read_csv("/nas/houce/UK_mobility/individual mobility profile previous/England_individual_1127.csv")
```

```python
data['home_days_2020_lad_mean'] = data.groupby('home_2020_lad_code')['home_days_2020'].transform('mean')
data['days_amenties_2020_all_lad_mean'] = data.groupby('home_2020_lad_code')['days_amenties_2020_all'].transform('mean')
data['workplace_days_2020_lad_mean'] = data.groupby('home_2020_lad_code')['workplace_days_2020'].transform('mean')

data['home_days_2020_norm'] = data['home_days_2020'] / data['home_days_2020_lad_mean']
data['days_amenties_2020_all_norm'] = data['days_amenties_2020_all'] / data['days_amenties_2020_all_lad_mean']
data['workplace_days_2020_norm'] = data['workplace_days_2020'] / data['workplace_days_2020_lad_mean']
```

```python
import matplotlib.pyplot as plt
import ternary
import numpy as np
import random

# ---------------------------------------------------------
# 1. 准备数据
# ---------------------------------------------------------
# 交换 'workplace_hours_2020' 和 'hours_amenties_2020_all' 的顺序
# 以匹配 ternary 库的坐标轴顺序 (bottom, right, left)
data_points = (data[['home_days_2020_norm', 'days_amenties_2020_all_norm', 'workplace_days_2020_norm']].dropna()).values

# ---------------------------------------------------------
# 2. 数据处理：网格计数
# ---------------------------------------------------------
scale = 100 
counts = dict()

for (a, b, c) in data_points:
    # 确保 a+b+c 大约等于 1
    total = a + b + c
    if total == 0: continue
    a, b, c = a/total, b/total, c/total # 归一化

    i = int(round(a * scale))
    j = int(round(b * scale))
    k = scale - i - j 
    
    if k < 0: continue 
    
    coord = (i, j, k)
    if coord in counts:
        counts[coord] += 1
    else:
        counts[coord] = 1

# ---------------------------------------------------------
# 3. 绘制三元热力图
# ---------------------------------------------------------
fig, tax = ternary.figure(scale=scale)
fig.set_size_inches(10, 8)

tax.heatmap(counts, style="triangular", cmap='YlGnBu', colorbar=True)

tax.boundary(linewidth=2.0)
tax.gridlines(color="black", multiple=10)
tax.gridlines(color="blue", multiple=1, linewidth=0.5, alpha=0.1)

fontsize = 12
tax.ticks(axis='lbr', linewidth=1, multiple=20, offset=0.025, tick_formats="%.1f")

tax.get_axes().axis('off')
tax.clear_matplotlib_ticks()
tax.set_title("Home, Workplace, Amenities Distribution (2020)", fontsize=18)

# 标签现在与数据轴匹配
tax.left_axis_label("Workplace Days (%)", offset=0.16)
tax.right_axis_label("Amenities Hours (%)", offset=0.16)
tax.bottom_axis_label("Home Hours (%)", offset=0.06)

tax.get_axes().set_facecolor('white')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import ternary
import pandas as pd
import numpy as np

# ==========================================
# 1. 数据准备 (模拟你的 DataFrame)
# ==========================================

df1 = data[['home_hours_2020', 'workplace_hours_2020', 'hours_amenties_2020_all']].dropna() / 24
df = pd.concat([df1, data['cluster_2020']], axis=1)

# ==========================================
# 2. 核心步骤：将属性值映射为颜色
# ==========================================
# 提取坐标列和属性列
points = df.iloc[:, 0:3].values
attribute_values = df.iloc[:, 3].values

# 定义颜色映射器 (Colormap)
# 可选: 'viridis', 'plasma', 'inferno', 'magma', 'cividis', 'CoolWarm', 'Blues' 等
cmap_name = 'tab10'
cmap = plt.get_cmap(cmap_name)

# 定义归一化器 (Normalize)
# 它的作用是将你的属性值 (比如 10到90) 线性映射到 0.0 到 1.0 之间
# vmin 和 vmax 定义了颜色条的最小值和最大值
norm = mcolors.Normalize(vmin=attribute_values.min(), vmax=attribute_values.max())

# 生成最终的颜色列表 (RGBA格式)
# 将归一化后的值传给 cmap，得到每个点对应的颜色
colors_mapped = cmap(norm(attribute_values))


# ==========================================
# 3. 绘制三元散点图
# ==========================================
# 设置图表比例尺，因为我们将数据乘以了100，所以 scale=100
scale = 100

# 创建画布和三元坐标轴
fig, tax = ternary.figure(scale=scale)
fig.set_size_inches(10, 8)

# --- 绘制散点 ---
# 使用我们映射好的颜色列表 colors_mapped
# s=80 设置点的大小, edgecolors='k' 设置点的黑色边框使其更清晰
tax.scatter(points, marker='o', color=colors_mapped, s=2, edgecolors='k', linewidth=0, zorder=3)

# --- 添加图表装饰 ---
# 绘制边界和网格线
tax.boundary(linewidth=2.0)
tax.gridlines(color="grey", multiple=10, linewidth=0.5) # 每隔10画一条网格线

# 设置轴标签和刻度
fontsize = 12
tax.set_title("Ternary Scatter Plot with Attribute Color mapping", fontsize=16, y=1.03)
# 设置三个角的标签
tax.left_axis_label("Component B (%)", fontsize=fontsize, offset=0.14)
tax.right_axis_label("Component C (%)", fontsize=fontsize, offset=0.14)
tax.bottom_axis_label("Component A (%)", fontsize=fontsize, offset=0.14)

# 设置刻度值
tax.ticks(axis='lbr', linewidth=1, multiple=10, offset=0.025, fontsize=10)

# 清除默认的 Matplotlib 坐标轴显示
tax.get_axes().axis('off')
tax.clear_matplotlib_ticks()

# ==========================================
# 4. 添加颜色条 (Colorbar)
# ==========================================
# 需要创建一个 "ScalarMappable" 对象来告诉 matplotlib 如何绘制颜色条
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([]) # 这一步是必须的，虽然传入的是空数组

# 在图的一侧添加颜色条
# shrink 控制颜色条的长度比例, aspect 控制宽度
cbar = plt.colorbar(sm, ax=tax.get_axes(), shrink=0.8, aspect=20, pad=0.05)
cbar.set_label('Attribute Value (Column 4)', fontsize=12, labelpad=10)

# 设置背景为白色
tax.get_axes().set_facecolor('white')

plt.show()
```
